```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single DataFrame
total_mouse_metadata = study_results.merge(mouse_metadata,on="Mouse ID",how="left")
total_mouse_metadata

# Display the data table for preview
```

```python
# Checking the number of mice.
total_mouse_metadata["Mouse ID"].nunique()
```

```python
# Our data should be uniquely identified by Mouse ID and Timepoint
# Get the duplicate mice by ID number that shows up for Mouse ID and Timepoint.
total_mouse_metadata.loc[total_mouse_metadata.duplicated(subset=["Timepoint","Mouse ID"],keep=False)]
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
total_mouse_metadata.loc[total_mouse_metadata["Mouse ID"]=="g989"]
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
mouse_ID_metadata=total_mouse_metadata.loc[total_mouse_metadata["Mouse ID"]!="g989"]
```

```python
# Checking the number of mice in the clean DataFrame.
mouse_ID_metadata["Mouse ID"].nunique()
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
# Assemble the resulting series into a single summary DataFrame.
means = mouse_ID_metadata.groupby('Drug Regimen')['Tumor Volume (mm3)'].mean()
medians = mouse_ID_metadata.groupby('Drug Regimen')['Tumor Volume (mm3)'].median()
variances = mouse_ID_metadata.groupby('Drug Regimen')['Tumor Volume (mm3)'].var()
sds = mouse_ID_metadata.groupby('Drug Regimen')['Tumor Volume (mm3)'].std()
sems = mouse_ID_metadata.groupby('Drug Regimen')['Tumor Volume (mm3)'].sem()

summary_table = pd.DataFrame({"Mean Tumor Volume":means,
                              "Median Tumor Volume":medians,
                              "Tumor Volume Variance":variances,
                              "Tumor Volume Std. Dev.":sds,
                              "Tumor Volume Std. Err.":sems})
summary_table
```

```python
# A more advanced method to generate a summary statistics table of mean, median, variance, standard deviation,
# and SEM of the tumor volume for each regimen (only one method is required in the solution)

# Using the aggregation method, produce the same summary statistics in a single line
summary_table = mouse_ID_metadata.groupby("Drug Regimen").agg({"Tumor Volume (mm3)":["mean","median","var","std","sem"]})
summary_table
```

## Line and Scatter Plots

```python
# Generate a line plot of tumor volume vs. time point for a single mouse treated with Capomulin
Cap_Mouse = mouse_ID_metadata.loc[mouse_ID_metadata['Drug Regimen'] == 'Capomulin','Mouse ID'].iloc[0]
Cap_Mouse_df = mouse_ID_metadata.loc[mouse_ID_metadata['Mouse ID'] == Cap_Mouse]
Cap_Mouse_df.plot(x = 'Timepoint', y = 'Tumor Volume (mm3)')
plt.show()
```

```python
# Generate a scatter plot of mouse weight vs. the average observed tumor volume for the entire Capomulin regimen
cap_mice = mouse_ID_metadata.loc[mouse_ID_metadata['Drug Regimen'] == 'Capomulin']
cap_mice_df = cap_mice.groupby('Mouse ID')[['Weight (g)', 'Tumor Volume (mm3)']].mean()
cap_mice_df.plot.scatter(x = 'Weight (g)', y = 'Tumor Volume (mm3)') 
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and a linear regression model 

weight = cap_mice_df["Weight (g)"]
tumor_volume = cap_mice_df["Tumor Volume (mm3)"]
correlation = st.pearsonr(weight, tumor_volume)
print(f"The correlation between both factors is {round(correlation[0], 2)}")

slope, intercept, r, p, se = st.linregress(cap_mice_df ['Weight (g)'], cap_mice_df['Tumor Volume (mm3)'])
slope, intercept, p

from scipy.stats import linregress
x_values = cap_mice_df['Weight (g)']
y_values = cap_mice_df['Tumor Volume (mm3)']
slope, intercept, rvalue, pvalue, stderr = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y=" + str(round(slope,2)) + "x + " + str(round(intercept,2))

plt.scatter(x_values,y_values)
plt.plot(x_values,regress_values,"r-")
plt.annotate(line_eq,(0,50),fontsize=15,color="red")
plt.xlabel('Weight')
plt.ylabel('Tumor Volume')
print(f"The r-squared is: {rvalue**2}")
plt.show()

# for mouse weight and average observed tumor volume for the entire Capomulin regimen
```
